```python
#Importing Libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("E://Datasets//New Dataset//Position_Salaries.csv")
```

```python
print(data.head(10))
```

```python
#Divide data into independent and dependent variable,
x_data=data.iloc[:,1:2].values
y_data=data.iloc[:,2].values
```

```python
x_data
```

```python
y_data
```

```python
#Importing random forest regressor.
from sklearn.ensemble import RandomForestRegressor
```

```python
#creating object 
reg=RandomForestRegressor(n_estimators=100,random_state=0)
```

```python
reg.fit(x_data,y_data)
```

```python
y_pred=reg.predict([[6.5]])
```

```python
print(y_pred)
```

```python
#Visualizing the Decision tree of random forest.
reg.estimators_
```

```python
from sklearn import tree
plt.figure(figsize=(15,20))
```

```python
tree.plot_tree(reg.estimators_[4],filled=True)
```
