```python
# importing basic library.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tree_data=pd.read_csv("E://Datasets//New Dataset//salaries.csv")
print(tree_data.head(10))
```

```python
# SPlitting data into dependent and Independent variable.
x_data=tree_data.iloc[:,0:3].values
y_data=tree_data.iloc[:,3].values
```

```python
x_data
```

```python
y_data
```

```python
#converting all label columns into number by using lable encoder.
from sklearn.preprocessing import LabelEncoder
```

```python
#creating object of labelEncoding.
l_company=LabelEncoder()
l_job=LabelEncoder()
l_degree=LabelEncoder()
```

```python
#Creating a separate column.
tree_data['company_new']=l_company.fit_transform(tree_data['company'])
tree_data['job_new']=l_job.fit_transform(tree_data['job'])
tree_data['degree_new']=l_degree.fit_transform(tree_data['degree'])
```

```python
tree_data
```

```python
#Dropping unwanted columns from dataset.
x_data=tree_data.drop(['company','job','degree','salary_more_then_100k'],axis=1)
print(x_data)
```

```python
y_data
```

```python
#importing decision tree libarry.
from sklearn import tree
```

```python
tree_reg=tree.DecisionTreeClassifier(random_state=0)
```

```python
tree_reg.fit(x_data,y_data)
```

```python
result=tree_reg.score(x_data,y_data)
```

```python
print(result)
```

```python
tree_reg.predict([[2,0,1]])
```

```python
#Visualize a decision tree.
plt.figure(figsize=(15,10))
tree.plot_tree(tree_reg,filled=True)
```
